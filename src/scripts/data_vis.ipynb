```python
import pandas as pd 
import sweetviz as sv 

import warnings
```

```python
train_df = pd.read_csv('../../data/data/train.csv')
val_df = pd.read_csv('../../data/feature/cb_validation.csv')
test_df = pd.read_csv('../../data/data/test.csv')
```

```python
train_col_report = sv.analyze(train_df)
train_col_report.show_html()
```

```python
val_data_report = sv.analyze(val_df)
val_data_report.show_html()
```

```python
test_data_report = sv.analyze(test_df)
test_data_report.show_html()
```

```python
tr_val_dist_report = sv.compare(train_df, val_df)
tr_val_dist_report.show_html()
```

```python
val_ts_dist_report = sv.compare(val_df,test_df)
val_ts_dist_report.show_html()
```
